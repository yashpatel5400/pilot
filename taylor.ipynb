```python
import jax.numpy as jnp
from jax import grad, vmap
```

```python
n = 10
f = jnp.sin
taylor = [f]

for i in range(n):
    taylor.append(grad(taylor[-1]))
```

```python
import matplotlib.pyplot as plt
import math
```

```python
x_0 = 0.0
def taylor_sin(x):
    return jnp.sum(jnp.array([taylor[k](x_0) / math.factorial(k) * (x - x_0) ** k for k in range(n)]), axis=0)
```

```python
x = jnp.arange(0, 5, .1)
y_approx = taylor_sin(x)
y = f(x)
```

```python
plt.plot(x, y)
plt.plot(x, y_approx)
```
